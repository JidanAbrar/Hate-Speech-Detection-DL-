```python
# [Cell 1] - Imports and Setup
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, accuracy_score, precision_score, recall_score, confusion_matrix, classification_report
from transformers import AutoTokenizer, AutoModelForSequenceClassification, AdamW  # Using Auto classes for flexibility
from torch.optim import lr_scheduler
from tqdm import tqdm
import random
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, HTML
import warnings
warnings.filterwarnings('ignore')

def set_seeds(seed=42):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True

set_seeds()

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# [Cell 2] - Dataset Class and Tokenizer Initialization
class HateSpeechDataset(Dataset):
    def __init__(self, texts, labels, tokenizer, max_length=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length
        
    def __len__(self):
        return len(self.texts)
    
    def __getitem__(self, idx):
        text = str(self.texts[idx])
        label = self.labels[idx]
        
        encoding = self.tokenizer(
            text,
            add_special_tokens=True,
            max_length=self.max_length,
            padding='max_length',
            truncation=True,
            return_attention_mask=True,
            return_tensors='pt'
        )
        
        return {
            'input_ids': encoding['input_ids'].flatten(),  # Removed negative sign (likely a typo)
            'attention_mask': encoding['attention_mask'].flatten(),
            'labels': torch.tensor(label, dtype=torch.long)
        }

# Initialize tokenizer for l3cube-pune/tamil-bert
try:
    tokenizer = AutoTokenizer.from_pretrained("mdosama39/tamil-bert-Caste-HateSpech_LTEDi-tamil")
    print("Loaded Tamil-BERT tokenizer successfully.")
except Exception as e:
    print(f"Error loading Tamil-BERT tokenizer: {e}")
    raise  # No fallback, as we specifically want Tamil-BERT

# [Cell 3] - Model Definition
def load_pretrained_model():
    label2id = {"Non-Hate": 0, "Hate": 1}
    id2label = {0: "Non-Hate", 1: "Hate"}
    
    try:
        model = AutoModelForSequenceClassification.from_pretrained(
            "mdosama39/tamil-bert-Caste-HateSpech_LTEDi-tamil",
            num_labels=2,
            problem_type="single_label_classification",
            id2label=id2label,
            label2id=label2id
        )
        print("Loaded Tamil-BERT model successfully.")
    except Exception as e:
        print(f"Error loading Tamil-BERT model: {e}")
        raise  # No fallback, as we specifically want Tamil-BERT
    
    return model.to(device)

# [Cell 4] - Focal Loss and Metrics Visualizer
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None):
        super(FocalLoss, self).__init__()
        self.gamma = gamma
        self.alpha = alpha if alpha is not None else torch.tensor([1.0, 1.0]).to(device)
    
    def forward(self, logits, targets):
        ce_loss = torch.nn.functional.cross_entropy(logits, targets, reduction='none')
        pt = torch.exp(-ce_loss)
        focal_loss = (self.alpha[targets] * (1 - pt) ** self.gamma * ce_loss).mean()
        return focal_loss

class MetricsVisualizer:
    def __init__(self):
        self.train_metrics = []
        self.val_metrics = []

    def add_metrics(self, train_metrics, val_metrics):
        self.train_metrics.append(train_metrics)
        self.val_metrics.append(val_metrics)

    def plot_metrics(self):
        epochs = range(1, len(self.train_metrics) + 1)
        metrics = ['loss', 'accuracy', 'f1', 'precision', 'recall']

        plt.figure(figsize=(15, 10))
        for i, metric in enumerate(metrics, 1):
            plt.subplot(2, 3, i)
            plt.plot(epochs, [m[metric] for m in self.train_metrics], 'b-', label=f'Train {metric}')
            plt.plot(epochs, [m[metric] for m in self.val_metrics], 'r-', label=f'Val {metric}')
            plt.title(f'{metric.capitalize()} over epochs')
            plt.xlabel('Epoch')
            plt.ylabel(metric.capitalize())
            plt.legend()
        plt.tight_layout()
        plt.show()

    def plot_confusion_matrix(self, y_true, y_pred):
        cm = confusion_matrix(y_true, y_pred)
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
        plt.title('Confusion Matrix')
        plt.ylabel('True Label')
        plt.xlabel('Predicted Label')
        plt.show()

# [Cell 5] - Training Function
def train_model(model, train_loader, val_loader, device, num_epochs=8, accum_steps=2):
    optimizer = AdamW(model.parameters(), lr=2e-5, weight_decay=0.05)
    scheduler = lr_scheduler.CosineAnnealingLR(optimizer, T_max=num_epochs)
    criterion = torch.nn.CrossEntropyLoss(weight=torch.tensor([1.0, 3415/2097]).to(device))
    metrics_viz = MetricsVisualizer()

    best_val_f1 = 0
    patience = 2
    patience_counter = 0
    best_model_path = "best_model.pt"

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        train_preds, train_labels = [], []
        optimizer.zero_grad()

        for i, batch in enumerate(tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs} - Training')):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = criterion(outputs.logits, labels) / accum_steps

            loss.backward()
            if (i + 1) % accum_steps == 0:
                torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
                optimizer.step()
                scheduler.step()
                optimizer.zero_grad()

            train_loss += loss.item() * accum_steps
            preds = torch.argmax(outputs.logits, dim=1).cpu().detach().numpy()
            train_preds.extend(preds)
            train_labels.extend(labels.cpu().numpy())

        train_metrics = {
            'loss': train_loss / len(train_loader),
            'accuracy': accuracy_score(train_labels, train_preds),
            'f1': f1_score(train_labels, train_preds, average='macro'),
            'precision': precision_score(train_labels, train_preds, average='macro'),
            'recall': recall_score(train_labels, train_preds, average='macro')
        }

        print(f"\nEpoch {epoch + 1} Training Metrics:")
        for metric, value in train_metrics.items():
            print(f"{metric}: {value:.4f}")

        model.eval()
        val_loss = 0
        val_preds, val_labels = [], []

        with torch.no_grad():
            for batch in tqdm(val_loader, desc=f'Epoch {epoch + 1}/{num_epochs} - Validation'):
                input_ids = batch['input_ids'].to(device)
                attention_mask = batch['attention_mask'].to(device)
                labels = batch['labels'].to(device)

                outputs = model(input_ids=input_ids, attention_mask=attention_mask)
                loss = criterion(outputs.logits, labels)

                val_loss += loss.item()
                preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
                val_preds.extend(preds)
                val_labels.extend(labels.cpu().numpy())

        val_metrics = {
            'loss': val_loss / len(val_loader),
            'accuracy': accuracy_score(val_labels, val_preds),
            'f1': f1_score(val_labels, val_preds, average='macro'),
            'precision': precision_score(val_labels, val_preds, average='macro'),
            'recall': recall_score(val_labels, val_preds, average='macro')
        }

        print(f"\nEpoch {epoch + 1} Validation Metrics:")
        for metric, value in val_metrics.items():
            print(f"{metric}: {value:.4f}")
        print(classification_report(val_labels, val_preds, target_names=['Non-Hate', 'Hate']))

        metrics_viz.add_metrics(train_metrics, val_metrics)

        if val_metrics['f1'] > best_val_f1:
            best_val_f1 = val_metrics['f1']
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'scheduler_state_dict': scheduler.state_dict(),
                'best_f1': best_val_f1,
            }, best_model_path)
            print(f"New best model saved with F1: {best_val_f1:.4f}")
            patience_counter = 0
        else:
            patience_counter += 1
            print(f"No improvement. Patience counter: {patience_counter}/{patience}")
            if patience_counter >= patience:
                print(f'Early stopping after {epoch + 1} epochs')
                break

    metrics_viz.plot_metrics()
    metrics_viz.plot_confusion_matrix(val_labels, val_preds)
    return model

# [Cell 6] - Data Preparation
def prepare_data(train_path, batch_size=16):
    data = pd.read_csv(train_path)
    if 'text' not in data.columns or 'label' not in data.columns:
        raise ValueError("Dataset must contain 'text' and 'label' columns")
    data = data.sample(frac=1).reset_index(drop=True)
    train_data, val_data = train_test_split(data, test_size=0.2, random_state=42, stratify=data['label'])
    
    print(f"Training: {len(train_data)}, Validation: {len(val_data)}")

    train_dataset = HateSpeechDataset(train_data['text'].values, train_data['label'].values, tokenizer)
    val_dataset = HateSpeechDataset(val_data['text'].values, val_data['label'].values, tokenizer)

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
    val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False, num_workers=2)

    return train_loader, val_loader, val_data

# [Cell 7] - Model Evaluation
def evaluate_model(model, val_loader, device):
    model.eval()
    criterion = torch.nn.CrossEntropyLoss(weight=torch.tensor([1.0, 3415/2097]).to(device))
    val_loss = 0
    all_preds, all_labels = [], []

    with torch.no_grad():
        for batch in tqdm(val_loader, desc='Evaluating'):
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            loss = criterion(outputs.logits, labels)

            val_loss += loss.item()
            preds = torch.argmax(outputs.logits, dim=1).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    metrics = {
        'loss': val_loss / len(val_loader),
        'accuracy': accuracy_score(all_labels, all_preds),
        'f1': f1_score(all_labels, all_preds, average='macro'),
        'precision': precision_score(all_labels, all_preds, average='macro'),
        'recall': recall_score(all_labels, all_preds, average='macro')
    }

    print("\nValidation Metrics (Final):")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")

    MetricsVisualizer().plot_confusion_matrix(all_labels, all_preds)
    return metrics, all_preds

# [Cell 8] - Main Execution
def main():
    train_path = "/kaggle/input/tamil-hate-speech/train (1).csv"
    train_loader, val_loader, val_data = prepare_data(train_path)

    model = load_pretrained_model()
    model = train_model(model, train_loader, val_loader, device, num_epochs=8, accum_steps=2)

    torch.save(model.state_dict(), "tamil_hate_speech_tamil_bert.pt")  # Updated filename
    print("Model saved successfully!")

    tokenizer.save_pretrained("./saved_tokenizer")
    print("Tokenizer saved successfully!")

    checkpoint = torch.load('best_model.pt')
    model.load_state_dict(checkpoint['model_state_dict'])
    model.to(device)

    val_metrics, val_preds = evaluate_model(model, val_loader, device)

    val_data['predictions'] = val_preds
    val_data.to_csv("val_predictions.csv", index=False)
    print("Validation predictions saved!")

if __name__ == "__main__":
    main()

# [Cell 9] - Error Analysis
def analyze_errors(df, actual_col='label', pred_col='predictions'):
    incorrect = df[df[actual_col] != df[pred_col]]
    print(f"Total incorrect predictions: {len(incorrect)}")
    display(incorrect.head(10))
    incorrect.to_csv("incorrect_predictions.csv", index=False)
    print("Incorrect predictions saved!")
    return incorrect

# Execute error analysis
if __name__ == "__main__":
    val_data = pd.read_csv("val_predictions.csv")
    incorrect_preds = analyze_errors(val_data)
```
